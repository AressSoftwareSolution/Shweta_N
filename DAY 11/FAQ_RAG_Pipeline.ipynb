```python
import pdfplumber

pdf_path = '/content/Aress_Task (5).pdf'
pdf_text = ''

with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        pdf_text += page.extract_text() + '\n'

print(f"test length : {len(pdf_text)}")
print(f"First 500 characters: \n{pdf_text[:500]}")
```

```python
import re

cleaned_pdf_text = pdf_text.replace('\n', ' ').strip()    ## remove next line
cleaned_pdf_text = re.sub(r'\s+', ' ', cleaned_pdf_text).lower()  ## remove extra space and lower case the text

print(f"cleaned text:\n{cleaned_pdf_text}")
```

```python
import string

# Remove punctuation
cleaned_pdf_text = cleaned_pdf_text.translate(str.maketrans('', '', string.punctuation))

print(f"Cleaned text length after punctuation removal: {len(cleaned_pdf_text)}")
print(f"cleaned text after punctuation removal:\n{cleaned_pdf_text}")
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python

# Tokenize the cleaned text
tokens = nltk.word_tokenize(cleaned_pdf_text)

print(f"Number of tokens: {len(tokens)}")
print(tokens)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word not in stop_words]  ## remove stop words

print(f"Number of tokens before stop word removal: {len(tokens)}")
print(f"Number of tokens after stop word removal: {len(filtered_tokens)}")
```

```python
print(filtered_tokens)
```

```python
final_processed_text = ' '.join(filtered_tokens)

print(f"Final processed text length: {len(final_processed_text)}")
print(f"{final_processed_text}")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Initialize the RecursiveCharacterTextSplitter
# Using a chunk_size of 500 and chunk_overlap of 50, these values can be adjusted.
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    length_function=len,
    add_start_index=True,
)

# Split the final_processed_text into chunks
text_chunks = text_splitter.split_text(final_processed_text)

print(f"Number of text chunks: {len(text_chunks)}")
print(f"First chunk:\n{text_chunks[0]}")
```

```python
from sentence_transformers import SentenceTransformer

# Initialize a SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings for each text chunk
text_embeddings = model.encode(text_chunks)

# Print the shape of the text_embeddings and the embedding of the first chunk
print(f"Shape of text embeddings: {text_embeddings.shape}")
print(f"Embedding of the first chunk:\n{text_embeddings[0]}")
```

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import SentenceTransformerEmbeddings
from langchain_core.documents import Document

# Initialize the SentenceTransformerEmbeddings for Chroma
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

# Convert text chunks to Document objects
document_chunks = [Document(page_content=t) for t in text_chunks]

# Create a Chroma vector store from the text chunks and embeddings
db = Chroma.from_documents(document_chunks, embeddings)

print("Chroma vector store created successfully.")
```

```python
retriever = db.as_retriever()

print("Retriever initialized from ChromaDB.")
```

```python
def retrieve_chunks(query, retriever, top_k=4):
    """
    Retrieves top-k relevant chunks from ChromaDB.
    """
    results = retriever.get_relevant_documents(query)
    return results[:top_k]
```

```python
def build_rag_prompt(query, retrieved_docs):
    """
    Builds RAG prompt for LLM using retrieved chunks.
    """
    context_text = "\n\n".join([f"Chunk {i+1}: {doc.page_content}"
                                for i, doc in enumerate(retrieved_docs)])

    prompt = f"""
You are an AI assistant for answering questions from the provided knowledge base.
Use ONLY the context below to answer the user query.
If answer is not present, reply "I’m only able to answer questions that relate to the FAQ realted to system only. If you have any questions on that topic, feel free to ask!".

### CONTEXT:
{context_text}

### USER QUESTION:
{query}

### RESPONSE:
"""
    return prompt
```

```python
from groq import Groq
from google.colab import userdata
```

```python

GROQ_API_KEY = userdata.get("GROK_API")

client = Groq(api_key=GROQ_API_KEY)
```

```python
def rag_query(query):
    retrieved_docs = retrieve_chunks(query, retriever)
    final_prompt = build_rag_prompt(query, retrieved_docs)
    response = client.chat.completions.create(
        model="openai/gpt-oss-120b",
        messages=[
            {"role": "system", "content": "You are a RAG-based assistant."},
            {"role": "user", "content": final_prompt},
        ],
        temperature=0.1
    )
    try:
        return response.choices[0].message.content
    except Exception as e:
        print("\n⚠ ERROR extracting content:", e)
        return None
```
