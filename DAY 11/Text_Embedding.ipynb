```python
from google import genai

search_query = input("Enter your search query:")

query_embedding_result = client.models.embed_content(
    model="gemini-embedding-001",
    contents=search_query
)
query_embedding = query_embedding_result.embeddings[0].values
```

```python
print(query_embedding)
```

```python
results = collection.query(
    query_embeddings=[query_embedding],
    n_results=1,
    include=['documents', 'distances']
)
print(results)
for i in range(len(results['documents'][0])):
    print(f"Document: Distance: {results['distances'][0][i]}")

    if results['distances'][0][i] >= 0.8:
      print("The Query is Not Related")
      break
    else:
      print("The Query is Related")
```
