```python
# Import libraries
from groq import Groq
from google.colab import userdata
```

```python
# Load API key from Colab's user secrets
GROQ_API_KEY = userdata.get("GROK_API")

# Initialize Groq client
client = Groq(api_key=GROQ_API_KEY)
```

```python
# Function to summarize text
def summarize_text(text):
    completion = client.chat.completions.create(
        model="openai/gpt-oss-120b",
        messages=[
            {
                "role": "user",
                "content": f"Summarize the following text in concise and clear English:\n\n{text}"
            }
        ],
        temperature=0.5,
        max_completion_tokens=1024,
        top_p=1,
        reasoning_effort="medium",
        stream=True
    )

    summary = ""

    # Stream the output
    for chunk in completion:
        if chunk.choices[0].delta.content:
            print(chunk.choices[0].delta.content, end="")
            summary += chunk.choices[0].delta.content

    return summary
```

```python
user_text = input("Enter text to summarize:\n\n")

print("\n\n---- SUMMARY ----\n")
summary_output = summarize_text(user_text)
```
