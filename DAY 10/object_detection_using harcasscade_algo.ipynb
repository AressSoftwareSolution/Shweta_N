```python
import cv2
import numpy as np
```

```python
def detect_and_crop_face(image_path):
    face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades+ 'haarcascade_frontalface_default.xml')
    image=cv2.imread(image_path)
    if image is None:
            print(f"Could not read the image at {image_path}")
            return None, None
    gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,1.3,5)
    if len(faces) == 0:
        print(f"No faces detected in the image at {image_path}")
        return None, image
    x,y,w,h=faces[0]
    cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)
    roi_color=image[y:y+h,x:x+w]
    return roi_color,image
```

```python
def preprocess_image(image):
    # image=cv2.imread(img_path)
    # if image is None:
    #     raise ValueError(f"Could not read the image at {img_path}")
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    image=cv2.resize(image,(640,640))
    image=image/255.0
    return image
```

```python
def process_image(image_path):
    crop_face,imagewith_box=detect_and_crop_face(image_path)
    preprocessed_image=preprocess_image(crop_face)
    return preprocessed_image,imagewith_box

face_image,boxed_image=process_image('/workspaces/Aress_Task/notebooks/PM.png')
# cv2.imshow("face Detected",boxed_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

import matplotlib.pyplot as plt
plt.imshow(cv2.cvtColor(boxed_image,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
