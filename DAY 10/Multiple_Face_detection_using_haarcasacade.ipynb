```python
## implement a simple computer vision model for image in which i want to fram the bounding box to the object in the image 


## using harcascassdes algorithm detecting the face in the image and then cropping the image to get the face region only. This is done to improve the performance of the model by reducing the background noise and focusing on the relevant features of the face.
import cv2

# Load Haar Cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

import cv2

# Load Haar Cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 
                                     "haarcascade_frontalface_default.xml")

def detect_multiple_faces(image_path):
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Could not read the image at {image_path}")

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect ALL faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5
    )

    # Draw bounding boxes for ALL faces
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x + w, y + h),
                      (0, 255, 0), 2)

    return image


boxed_image=detect_multiple_faces('/workspaces/Aress_Task/notebooks/OIP.png')

import matplotlib.pyplot as plt

plt.imshow(cv2.cvtColor(boxed_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Faces Detected")
plt.show()
```
