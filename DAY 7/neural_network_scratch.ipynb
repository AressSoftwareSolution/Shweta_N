```python
weight=1
bias=0
input_1=[1,0,1,0]
thresold=2
```

```python
# Calculate the Weighted Sum for ML neuron model
weighted_sum_1 = sum(w * x for w, x in zip([weight] * len(input_1), input_1)) + bias
```

```python
if weighted_sum_1>= thresold:
    print("Output is True ")
else:
    print("Output is false")
```
